# Words

fname = "新版HSK考试大纲-词汇.txt"

```python
with open(fname, "r", encoding="utf-8") as f:
    data = f.read().strip().split("\n")

text = ""
for line in data:
    if "序号	等级	词语	拼音	词性" in line or line.strip() == "":
        continue
    l = line.split("\t")
    if len(l) >= 4:
        text += "\t".join(l[:]).strip() + "\n"
```

```python
with open("新版HSK考试大纲-词汇_cleaned.txt", "w", encoding="utf-8") as f:
    f.write(text)
```

```python
fname = "新版HSK考试大纲-词汇_cleaned.txt"

words = {"1": [], "2": [], "3": [], "4": [], "5": [], "6": [], "7-9": []}

import re

with open(fname, "r", encoding="utf-8") as f:
    data = f.read().strip().split("\n")
    for line in data:
        # print(line)
        l = line.split("\t")
        count = l[0]
        level = l[1]
        word = l[2]
        pinyin = l[3]
        pos = l[4] if len(l) > 4 else ""
        # print(f"Level: {level}, Word: {word}, Pinyin: {pinyin}, POS: {pos}")

        if "（" in level:
            nums = re.findall(r"\d+(?:-\d+)?", level)
            for lev in nums:
                if lev not in words:
                    words[lev] = []
                words[lev].append((count, level, word, pinyin, pos))
        else:
            words[level].append((count, level, word, pinyin, pos))
```

```python
for level in words:
    print(f"Level {level}: {len(words[level])} words")
    with open(f"../New HSK (2025)/HSK Words/temp/HSK_Level_{level}_words.txt", "w", encoding="utf-8") as f:
        for count, level, word, pinyin, pos in words[level]:
            f.write(f"{count}\t{level}\t{word}\t{pinyin}\t{pos}\n")
```

# Hanzi

```python
fdir = "../New HSK (2025)/HSK Hanzi/temp"

hanzi = {"1": [], "2": [], "3": [], "4": [], "5": [], "6": [], "7-9": []}

def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

for level in hanzi:
    with open(f"{fdir}/HSK {level} Hanzi.txt", "r", encoding="utf-8") as f:
        data = f.read().strip().split("\n")
        for line in data:
            l = line.replace(".", "\t")
            print(l)
            for item in l.split("\t"):
                item = item.strip()
                if item == "" or has_numbers(item):
                    continue
                hanzi[level].append(item)
        
```

```python
for level in hanzi:
    print(f"Level {level}: {len(hanzi[level])} hanzi")
    with open(f"../New HSK (2025)/HSK Hanzi/HSK_Level_{level}_hanzi.txt", "w", encoding="utf-8") as f:
        for h in hanzi[level]:
            f.write(f"{h}\n")
```
